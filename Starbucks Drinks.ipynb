## Import Libraries

```python
# Import necessary libraries
import pandas as pd  # For data manipulation and analysis
import seaborn as sns # For statistical data visualization
import matplotlib.pyplot as plt # For creating static, animated, and interactive visualizations
```

## Reading Data

```python
# Load the Starbucks drinks dataset into a DataFrame
starbucks_df = pd.read_csv("starbucks_drinks.csv")  # Read CSV file into DataFrame
starbucks_df  # Display the DataFrame to inspect the data
```

## Understanding Data

```python
# Display information about the DataFrame
starbucks_df.info()  # Shows the number of entries, column names, data types, and non-null counts
```

```python
# Clean up column names by replacing spaces with underscores and stripping trailing underscores
starbucks_df.columns = starbucks_df.columns.str.replace(" ", "_").str.strip("_")  # Modify column names
starbucks_df.columns  # Display the cleaned column names
```

```python
starbucks_df[starbucks_df['Caffeine_(mg)'] == 'Varies']
```

```python
starbucks_df = starbucks_df.drop(starbucks_df[starbucks_df['Caffeine_(mg)'] == 'varies'].index).reset_index(drop=True)
starbucks_df
```

```python
# Remove percentage symbols from nutrient columns
starbucks_df[['Vitamin_A_(%_DV)', 'Vitamin_C_(%_DV)', 'Calcium_(%_DV)', 'Iron_(%_DV)']] = starbucks_df[['Vitamin_A_(%_DV)', 'Vitamin_C_(%_DV)', 'Calcium_(%_DV)', 'Iron_(%_DV)']].replace('%', '', regex=True)  # Replace '%' with empty string
starbucks_df  # Display the updated DataFrame
```

```python
# Convert percentage columns to float and normalize by dividing by 100
starbucks_df[['Vitamin_A_(%_DV)', 'Vitamin_C_(%_DV)', 'Calcium_(%_DV)', 'Iron_(%_DV)']] = starbucks_df[['Vitamin_A_(%_DV)', 'Vitamin_C_(%_DV)', 'Calcium_(%_DV)', 'Iron_(%_DV)']].astype("float64")/100

# Convert columns to numeric, coercing errors (e.g., invalid parsing) and fill NaNs
starbucks_df['Total_Fat_(g)'] = pd.to_numeric(starbucks_df['Total_Fat_(g)'], errors='coerce')  # Convert to numeric
starbucks_df['Caffeine_(mg)'] = pd.to_numeric(starbucks_df['Caffeine_(mg)'], errors='coerce')  # Convert to numeric

# Fill NaN values with 0 and convert data types
starbucks_df['Total_Fat_(g)'] = starbucks_df['Total_Fat_(g)'].fillna(0).astype('float64')  # Replace NaNs with 0 and convert to float
starbucks_df['Caffeine_(mg)'] = starbucks_df['Caffeine_(mg)'].fillna(0).astype('int64')  # Replace NaNs with 0 and convert to int

```

## EDA 

# Univariate Analysis

```python
# Univariate Analysis for Categorical Columns
for col in starbucks_df.select_dtypes(include=['object']).columns:
    print(f"Column: {col}")
    print(starbucks_df[col].value_counts())
    
    # Plotting Bar Plot for Categorical Data
    plt.figure(figsize=(10, 6))
    sns.countplot(data=starbucks_df, x=col, palette='viridis')
    plt.title(f"Distribution of {col}")
    plt.xticks(rotation=90)
    plt.show()

# Univariate Analysis for Numerical Columns
for col in starbucks_df.select_dtypes(include=['number']).columns:
    print(f"Column: {col}")
    print(starbucks_df[col].describe())
    
    # Plotting Histogram
    plt.figure(figsize=(10, 6))
    sns.histplot(starbucks_df[col], kde=True)
    plt.title(f"Distribution of {col}")
    plt.show()
    
    # Plotting Box Plot
    plt.figure(figsize=(10, 6))
    sns.boxplot(y=starbucks_df[col])
    plt.title(f"Box Plot of {col}")
    plt.show()

# Correlation Heatmap for Numerical Columns
numerical_data = starbucks_df.select_dtypes(include=['number'])
plt.figure(figsize=(12, 8))
sns.heatmap(numerical_data.corr(), annot=True, cmap='viridis', linewidths=0.5)
plt.title('Correlation Heatmap of Numerical Features')
plt.show()
```

# Business questions

#### 1) How do caffeine counts vary across different beverage categories?


```python
# Group by 'Beverage_category' and count the number of entries for 'Caffeine_(mg)'
starbucks_df.groupby("Beverage_category")['Caffeine_(mg)'].count().reset_index().sort_values(by='Caffeine_(mg)',ascending=False).reset_index(drop=True)
```

```python
# Create a box plot for caffeine content by beverage category
plt.boxplot(
    [starbucks_df[starbucks_df['Beverage_category'] == cat]['Caffeine_(mg)'] for cat in starbucks_df['Beverage_category'].unique()]
)
plt.xticks(
    range(1, len(starbucks_df['Beverage_category'].unique()) + 1),  # Set x-ticks to match the number of categories
    starbucks_df['Beverage_category'].unique()  # Label x-ticks with beverage categories
)
plt.xticks(rotation=90)  # Rotate x-tick labels for better readability
plt.xlabel('Beverage Category')  # Label x-axis
plt.ylabel('Caffeine (mg)')  # Label y-axis
plt.title('Caffeine Content by Beverage Category')  # Set plot title
plt.show()  # Display the plot
```

#### 2) How do different beverage categories compare in terms of calories, fat and sugar content?

```python
# Calculate the percentage of total calories contributed by each beverage category
calories_percentage = round(
    (starbucks_df.groupby("Beverage_category")["Calories"].sum() / starbucks_df["Calories"].sum() * 100)  # Compute percentage
      # Round to 1 decimal place
).reset_index().sort_values(by='Calories',ascending=False).reset_index(drop=True)
calories_percentage  # Display the resulting DataFrame
```

```python
total_fat_percentage = round(((starbucks_df.groupby("Beverage_category")["Total_Fat_(g)"].sum())/(starbucks_df["Total_Fat_(g)"].sum())*100)
                             ,1).reset_index().sort_values(by='Total_Fat_(g)',ascending=False).reset_index(drop=True)
total_fat_percentage
```

```python
# Calculate the percentage of sugars for each beverage category
sugars_percentage = round(((starbucks_df.groupby("Beverage_category")["Sugars_(g)"].sum())/(starbucks_df["Sugars_(g)"].sum())*100)
                          ,1).reset_index().sort_values(by='Sugars_(g)',ascending=False).reset_index(drop=True)

# Display the result
sugars_percentage
```

```python
# Create subplots with a specific figure size for pie charts
fig, axes = plt.subplots(2, 2, figsize=(18, 12))

# Plot pie chart for percentage of calories by beverage category
plt.subplot(2, 2, 1)
plt.pie(
    calories_percentage["Calories"],  # Data for pie chart
    labels=calories_percentage['Beverage_category'],  # Labels for each slice
    autopct="%1.1f%%"  # Display percentage on the pie chart
)
plt.title('Percentage of Calories by Beverage Category')  # Title for the chart

# Plot pie chart for percentage of total fat by beverage category
plt.subplot(2, 2, 2)
plt.pie(
    total_fat_percentage["Total_Fat_(g)"],  # Data for pie chart
    labels=total_fat_percentage['Beverage_category'],  # Labels for each slice
    autopct="%1.1f%%"  # Display percentage on the pie chart
)
plt.title('Percentage of Total Fat by Beverage Category')  # Title for the chart

# Plot pie chart for percentage of sugars by beverage category
plt.subplot(2, 2, 3)
plt.pie(
    sugars_percentage["Sugars_(g)"],  # Data for pie chart
    labels=sugars_percentage['Beverage_category'],  # Labels for each slice
    autopct="%1.1f%%"  # Display percentage on the pie chart
)
plt.title('Percentage of Sugars by Beverage Category')  # Title for the chart

# Placeholder subplot with no pie chart
plt.subplot(2, 2, 4)
plt.axis('off')  # Hide this subplot

# Adjust layout to prevent overlap
plt.tight_layout()

# Display the plots
plt.show()
```

#### 4) What is the impact of different beverage preparations on the vitamin A , C , Calcium and Iron content?

```python
# Group by 'Beverage_prep' and calculate the mean values of selected nutrients
nutrient_averages = starbucks_df.groupby('Beverage_prep')[['Vitamin_A_(%_DV)', 'Vitamin_C_(%_DV)', 'Calcium_(%_DV)', 'Iron_(%_DV)']].mean().reset_index()

# Sort the results by 'Beverage_prep'
nutrient_averages_sorted = nutrient_averages.sort_values(by='Beverage_prep')

# Display the sorted DataFrame
nutrient_averages_sorted
```

```python
# Calculate the mean values for selected vitamins and minerals grouped by 'Beverage_prep'
vitamin_mineral_means = starbucks_df.groupby('Beverage_prep')[['Vitamin_A_(%_DV)', 'Vitamin_C_(%_DV)', 'Calcium_(%_DV)', 'Iron_(%_DV)']].mean().reset_index()

# Melt the DataFrame to long format for easier plotting
vitamin_mineral_melted = vitamin_mineral_means.melt(id_vars='Beverage_prep', var_name='Vitamin/Mineral', value_name='Mean (%DV)')

# Create a bar plot for the mean vitamin and mineral content by beverage preparation
plt.figure(figsize=(10, 6))
sns.barplot(
    x='Beverage_prep',  # X-axis: Beverage preparation types
    y='Mean (%DV)',  # Y-axis: Mean percentage of daily value
    hue='Vitamin/Mineral',  # Different bars for each vitamin/mineral
    data=vitamin_mineral_melted,  # Data for plotting
    palette='viridis'  # Color palette for the plot
)
plt.xlabel('Beverage Preparation')  # Label for the x-axis
plt.ylabel('Mean (%DV)')  # Label for the y-axis
plt.title('Mean Vitamin and Mineral Content by Beverage Preparation')  # Title of the plot
plt.legend(title='Vitamin/Mineral')  # Legend title
plt.xticks(rotation=90)  # Rotate x-tick labels for better readability
plt.show()  # Display the plot
```
